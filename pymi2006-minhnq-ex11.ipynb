V·∫Ω ƒë·ªì th·ªã h√†m sin,cos,tan trong kho·∫£ng -2ùúã -> 2ùúã


```python
import math
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-math.pi * 2, math.pi * 2, num=20)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.tan(x))

plt.title("Sin Cos Tan")
plt.legend(['y = sin(x)', 'y = cos(x)', 'y = tan(x)'])

```
